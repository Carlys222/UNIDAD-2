HipÃ³tesis nula (Hâ‚€):
Las variables masa corporal, media cintura y peso no tienen efecto significativo sobre el riesgo de hipertensiÃ³n.

HipÃ³tesis alternativa (Hâ‚):
Al menos una de las variables (masa corporal, media cintura o peso) tiene un efecto significativo sobre el riesgo de hipertensiÃ³n.

"El aumento en la masa corporal, la media de la cintura o el peso estÃ¡ asociado con un mayor riesgo de hipertensiÃ³n."

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['masa_corporal', 'medida_cintura', 'peso']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)


from scipy.stats import pearsonr
r,_ = pearsonr(Y_test, Yc)
print(f'Coeficiente de correlacion: {r: 0.4f}')
```

Los resultados del modelo de regresiÃ³n muestran que la masa corporal, medida de cintura y peso tienen una relaciÃ³n moderada con la hipertensiÃ³n arterial (r â‰ˆ 0.4856). Esto indica que estas variables influyen en el desarrollo de la hipertensiÃ³n, pero no son suficientes por sÃ­ solas para predecirla con alta precisiÃ³n. Por lo tanto, es probable que otros factores tambiÃ©n estÃ©n involucrados y deban considerarse para mejorar el modelo.

```python
r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```

El coeficiente de determinaciÃ³n es de
 =0.2339 sugiere que solo el 23.39% de la variabilidad en la hipertensiÃ³n puede explicarse mediante estas tres variables. Esto significa que, aunque hay cierta relaciÃ³n, la mayor parte de los factores que influyen en la hipertensiÃ³n no estÃ¡n incluidos en este modelo, por lo que serÃ­a necesario incorporar otras variables clÃ­nicas, genÃ©ticas o de estilo de vida para obtener una predicciÃ³n mÃ¡s precisa.

```python
b0, b1, b2, b3 = modelo.params

Fun = lambda X1, X2, X3: b0 + b1 * X1 + b2 * X2 + b3 * X3

Fun(30, 60, 120)
```

El modelo estima que una persona con esos valores fÃ­sicos tiene una probabilidad aproximada del 57.61% de presentar hipertensiÃ³n arterial.

Esto indica un riesgo moderado a elevado, segÃºn los datos usados para entrenar el modelo.

SUPUESTOS

```python
import matplotlib.pyplot as plt

# Suponiendo que ya tienes:
# Y_test: valores reales
# Yc: valores predichos por el modelo lineal

# Calcular residuos
residuos = Y_test - Yc

# GrÃ¡fico de dispersiÃ³n: residuos vs valores predichos
plt.scatter(Yc, residuos)
plt.axhline(y=0, color='red', linestyle='--')  # LÃ­nea en y=0
plt.xlabel('Valores predichos')
plt.ylabel('Residuos')
plt.title('VerificaciÃ³n de Linealidad: Residuos vs Valores Predichos')
plt.show()
```

No se cumple la suposiciÃ³n de linealidad; se observa un patrÃ³n sistemÃ¡tico en los residuos.

```python
from scipy.stats import shapiro
_, valor_p_shapiro = shapiro(modelo.resid)
print(f'Valor-p de Shapiro-Wilk: {valor_p_shapiro: 0.4f}\n')
```

Esto indica que los residuos del modelo no siguen una distribuciÃ³n normal, por lo que se viola el supuesto de normalidad en la regresiÃ³n lineal.

Esta falta de normalidad puede afectar la confiabilidad de las inferencias estadÃ­sticas del modelo, como los intervalos de confianza y las pruebas de hipÃ³tesis.

```python
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(modelo.resid, X_train_constante)
print(f'Valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}')
```

HipÃ³tesis nula (Hâ‚€): Los errores del modelo son homocedÃ¡sticos, es decir, tienen una varianza constante.

HipÃ³tesis alternativa (Hâ‚): Los errores del modelo son heterocedÃ¡sticos, es decir, la varianza de los errores varÃ­a a lo largo de las observaciones.

El valor-p de Breusch-Pagan de 0.0000 indica que hay heterocedasticidad significativa en tu modelo, lo que significa que la varianza de los errores no es constante

```python
from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'riesgo_hipertension ~ edad + masa_corporal + tension_arterial', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)  # ANOVA de una via
tabla_anova
```

El modelo muestra que de las tres variables que se usaron influyen de manera significativa en la variable dependiente, sienda masa corporal el factor mas determinante.
